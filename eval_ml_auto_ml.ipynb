```python
import evalml
X, y = evalml.demos.load_breast_cancer()
X_train, X_test, y_train, y_test = evalml.preprocessing.split_data(X, y, problem_type='binary')
```

```python
X_train.head()
```

```python
evalml.problem_types.ProblemTypes.all_problem_types
```

```python
from evalml.automl import AutoMLSearch
automl = AutoMLSearch(X_train = X_train, y_train = y_train, problem_type='binary', )
```

```python
automl.search()
```

```python
automl.rankings
```

```python
# Get best pipeline
automl.best_pipeline
```

```python
best_pipeline = automl.best_pipeline
```

```python
# Detail description of the model
automl.describe_pipeline(automl.rankings.iloc[0]["id"])
```

```python
best_pipeline.score(X_test, y_test, objectives=['auc','Precision','f1', 'Recall'])
```

```python
# Hyperparameter tune for specific objective like auc and so on
automl = AutoMLSearch(X_train = X_train, y_train = y_train, problem_type='binary', objective='auc', additional_objectives=['f1','precision'])
automl.search()
```

```python
automl.rankings
```

```python
best_pipeline = automl.best_pipeline
best_pipeline.score(X_test, y_test, objectives=['auc','Precision','f1', 'Recall'])
```

```python
automl.describe_pipeline(automl.rankings.iloc[0]["id"])
```

```python
# Save model object as pickle
best_pipeline.save('automl.pkl')
```

```python
# Load model
model = automl.load('automl.pkl')
```

```python
model.predict_proba(X_test).to_dataframe()
```
